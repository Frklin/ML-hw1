```python
#importing libraries
import numpy as np
import pandas as pd
import random

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import *
from sklearn.naive_bayes import *
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv("train_set.tsv", sep='\t', header=0)
#normalizing the dataset with min-max normalization

X = dataset.iloc[:,:-2]
y = dataset.iloc[:,-1]

#normalizing the dataset with min-max normalization
X = (X - X.min())/(X.max() - X.min())
y = (y - y.min())/(y.max() - y.min())

print('File loaded: %d samples.' %(len(dataset)))
```

```python
#regression with SVM
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

#splitting the dataset into the training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#fitting the regression model to the dataset
regressor = SVR(kernel = 'poly', gamma=0.3, degree=2, C=0.3)
regressor.fit(X_train, y_train)

#predicting the test set results
y_pred = regressor.predict(X_test)

#evaluating the model performance
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

#visualizing the results

plt.plot(np.arange(1,len(y_test)+1), y_test, color='red')
plt.plot(np.arange(1,len(y_test)+1), y_pred, color='blue')

```

```python
#regression with Random Forest
from sklearn.ensemble import RandomForestRegressor

#training the model
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)

#predicting the test set results
y_pred = regressor.predict(X_test)

#evaluating the model performance
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

print(len(y_test))
#plotting the results
plt.plot(np.arange(1,len(y_test)+1), y_test, color='red')
plt.plot(np.arange(1,len(y_test)+1), y_pred, color='blue')
#plotting the results
# plt.scatter(y_test, y_pred, color='red')
# plt.plot(y_test, y_test, color='blue')
# plt.title('Regression with Random Forest')
# plt.xlabel('Actual')
# plt.ylabel('Predicted')
# plt.show()

```

```python
#regression with Decision Tree
from sklearn.tree import DecisionTreeRegressor

#training the model
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train, y_train)

#predicting the test set results
y_pred = regressor.predict(X_test)

#evaluating the model performance
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

#plotting the results
plt.plot(np.arange(1,len(y_test)+1), y_test, color='red')
plt.plot(np.arange(1,len(y_test)+1), y_pred, color='blue')
plt.show()
```

```python
#regression with svm
from sklearn.svm import SVR

#training the model
regressor = SVR(kernel='rbf')
regressor.fit(X_train, y_train)

#predicting the test set results
y_pred = regressor.predict(X_test)

#evaluating the model performance
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

#plotting the results
plt.plot(np.arange(1,len(y_test)+1), y_test, color='red')
plt.plot(np.arange(1,len(y_test)+1), y_pred, color='blue')
plt.show()
```
